```
import numpy as np
import pandas as pd
from nolds.measures import corr_dim
from google.colab import drive
drive.mount('/content/drive')
!ls drive/MyDrive/Colab_data/*
import matplotlib.pyplot as plt

```

```
data = pd.read_csv('drive/MyDrive/Colab_data/GDAX.BTC-USD_200215_210215.csv')
data.drop(['<TICKER>', '<PER>'], axis=1, inplace=True)
data['<TIME>'] = data['<TIME>'].astype(str).str.zfill(6)
data['DATE'] = data[['<DATE>', '<TIME>']].apply(lambda x: ''.join(x.astype(str)), axis=1)
data.drop(['<DATE>', '<TIME>'], axis=1, inplace=True)
data['DATE'] = pd.to_datetime(data['DATE'].astype(int), format='%Y%m%d%H%M%S')
data.set_index("DATE")
data.rename(columns={'<OPEN>':'Open',
                     '<HIGH>':'High',
                     '<LOW>':'Low',
                     '<CLOSE>':'Close',
                     '<VOL>':'Volume',
                     'DATE':'Timestamp'}, inplace=True)
data.set_index('date')
```

```
high_s = data['high'][:15000].copy()
high_s.index = data['date'][:15000]
high_s
```

```
fractal_design = corr_dim(high_s, 5)
```
