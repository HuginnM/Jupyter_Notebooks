#### 1. Implement function str_to_dict(some_str) which returnes dictionary, where keys are string characters, and values are their quantity in the string:

<div style="text-align: right">1 point</div>

```python
def str_to_dict(some_str):  
    """
    :param some_str: str
    :return: dict
    """
    d = {}
    for now in some_str:
        d[now] = d.get(now, 0) + 1

    return d

    # YOUR CODE HERE
```

<div style="text-align: right">1 point
  

```python
def sec_smallest(numbers):
    """
    :param numbers: list[int]
    :return: int    
    """
    s = list(set(numbers))
    s.remove(min(s))

    return min(s)
    # YOUR CODE HERE
```

#### 3. Implement function prime_nums(n) that returns list of numbers which  are simple and < n:

<div style="text-align: right">1 point</div>

```python
def prime_nums(n):
    """
    :param n: int
    :return: list[int]
    """
    # YOUR CODE HERE
    result = []

    for i in range(2, n):
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            result.append(i)

    return result
```

**5. Implement function gcd(x, y), which returns the greatest common divisor of n and m.**
<div style="text-align: right">1 point</div>

```python
def gcd(x, y):   
    '''
    :params m,n: int
    :return: int
    '''
    # YOUR CODE HERE
    div_x = set()
    div_y = set()
    for divisor in range(1, x):
        if x % divisor == 0:
            div_x.add(divisor)

    for divisor in range(1, y):
        if y % divisor == 0:
            div_y.add(divisor)

    return max(div_x & div_y)
```

#### 6. Implement recursive sum of the list:
<div style="text-align: right">1 point</div>

```python
def recursive_list_sum(data_list):
    """
    :param data_list: list[list]
    """
    # YOUR CODE HERE
    if len(data_list) == 0:
        return 0
    if type(data_list[-1]) == list:
        temp = sum(data_list[-1])
        data_list.pop()
        return temp + recursive_list_sum(data_list)
    else:
        temp = data_list[-1]
        data_list.pop()
        return temp + recursive_list_sum(data_list)
```

#### 7. Implement decorator which returns function signature and it's return value:
<div style="text-align: right">1 point</div>

```python
def debug(func):
    """
    :param func: function
    """
    # YOUR CODE HERE
    def dec(*args):
        message = str(func.__name__) + str(args) + ' was called and returned ' + str(func(*args))
        return message
    return dec
```

#### 8. Implement class Conv, that contains method to_roman(self, n), which converts decimal numbers to Roman numerals:

<div style="text-align: right">2 points</div>

```python
class Conv:
    def __init__(self):
        self.val = [
            1000, 900, 500, 400,
            100, 90, 50, 40,
            10, 9, 5, 4, 1
        ]

        self.syb = [
            'M', 'CM', 'D', 'CD',
            'C', 'XC', 'L', 'XL',
            'X', 'IX', 'V', 'IV',
            'I'
        ]

    def to_roman(self, num):    
        """
        :param self:
        :param n: int
        :return: str    
        """
        # YOUR CODE HERE
        first_digit = num%10
        second_digit = num%100 // 10
        third_digit = num%1000 // 100
        fourth_digit = num%10000 // 1000
        
        rom_num = [fourth_digit * self.syb[0]]

        if third_digit * 100 in self.val:
            rom_num.append(self.syb[self.val.index(third_digit * 100)])
        elif third_digit in (2, 3):
            rom_num.append(self.syb[4]*third_digit)
        elif third_digit in (6, 7, 8):
            rom_num.append(self.syb[2] + self.syb[4]*third_digit)
        
        if second_digit*10 in self.val:
            rom_num.append(self.syb[self.val.index(second_digit * 10)])
        elif second_digit in (2, 3):
            rom_num.append(self.syb[8] * second_digit)
        elif second_digit in (6, 7, 8):
            rom_num.append(self.syb[6] + self.syb[8] * second_digit)
        
        if first_digit in self.val:
            rom_num.append(self.syb[self.val.index(first_digit)])
        elif first_digit in (2, 3):
            rom_num.append(self.syb[-1] * first_digit)
        elif first_digit in (6, 7, 8):
            rom_num.append(self.syb[-3] + self.syb[-1] * second_digit)
          
        return (''.join(rom_num))
```

#### 9. Implement class CombinationsList, that contains method get_combinations(self, my_list), which returns all combinations of elements of given list, including empty element and in order, law of which can be discrovered from the expected output:
<div style="text-align: right">3 points</div>

```python
class CombinationsList:
    @staticmethod
    def get_combinations(iterable):
        """
        :param self:
        :param my_list: list
        :return: list[list]
            """
            # YOUR CODE HERE
        def combinations(iterable, r):
            res = []
            pool = tuple(iterable)
            n = len(pool)
            if r > n:
                return
            indices = list(range(r))
            res.append([pool[i] for i in indices])
            while True:
                for i in reversed(range(r)):
                    if indices[i] != i + n - r:
                        break
                else:
                    return res
                indices[i] += 1
                for j in range(i+1, r):
                    indices[j] = indices[j-1] + 1
                res.append([pool[i] for i in indices])
            return res

        res = [[]]*(2**(len(iterable)))
        k = 0
        for j in range(len(iterable)+1):
            for comb in combinations(iterable, j):
                res[k] = comb
                k += 1
        return res

```

#### 10. Create base class <code>Rocket</code> and it's subclass <code>Shuttle</code>. The parent class contains methods <code>getMission()</code>, <code>addMission()</code> and <code>getName()</code>. The subclass <code>Shuttle</code> contains method <code>getDescription()</code>. Notice that we have not defined <code>getName()</code>, <code>getMission()</code>, <code>addMission()</code> in the <code>Shuttle</code> class but we are still able to access them, because the class <code>Shuttle</code> inherits them from the <code>Rocket</code> class. Use <code>super()</code> method to have ability to call those methods of the base class.
<div style="text-align: right">3 points</div>

```python
class Rocket:

    def __init__(self, name, mission):
        """
        :param name: str
        :param mission: str or list
        """
        # attributes are private to class Rocket
        self.__name = name      
        self.__mission = mission

    def getMission(self): 
        """
        : return str or list
        """
        # YOUR CODE HERE
        return self.__mission

    def addMission(self, mission): 
        # procedure method which adds a new mission. There can be one (str) or multiple (list) existing missions
        """
        : param mission: str
        """
        # YOUR CODE HERE
        r = [self.__mission]
        r.append(mission)
        self.__mission = r

    def getName(self):
        """
        : return str
        """
        # YOUR CODE HERE
        return self.__name

    
class Shuttle(Rocket):

    def __init__(self, name, mission, model):
        # call parent constructor to set name and mission  
        """
        :param name: str
        :param mission: str or list
        : param model: str
        """
        # YOUR CODE HERE
        super().__init__(name, mission)
        self.__model = model
        
    def getDescription(self):
        return 'Name: {0}\nModel: {1}\nMissions: {2}'.format(self.getName(), self.__model, str(self.getMission()))
```

```python
dragon = Shuttle("Crew Dragon", "Dragon 2 pad abort test", "V2")
print(dragon.getDescription(), '\n')
dragon.addMission('Dragon 2 in-flight abort test')
print(dragon.getDescription())
```
